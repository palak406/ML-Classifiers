<h1>Linear Regression Models using Housing Dataset</h1>

```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from sklearn import metrics
import seaborn as sns
```

```python
#Reading data
Housing_data = pd.read_csv(r'C:\Users\palak\Desktop\Housing.csv')
```

```python
Housing_data
```

```python
#I'm using heatmap for the columns to check the correlation.
fig_plt = plt.gcf() 
fig_size = fig_plt.get_size_inches()
fig_plt.set_size_inches(fig_size * 2.5)
sns.heatmap(Housing_data.corr(),annot=True)
plt.show()
```

```python
#From the heatmap we can say that TAx is highly correlated with the RAD,NOX and INDUS and therefore we would be using them as features and target variable. 
X=Housing_data[['RAD','NOX','INDUS']]
y=Housing_data['TAX']
```

```python
#Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
#Using Linear Regression
linear=LinearRegression()
linear.fit(X_train,y_train)
predicted_linear = linear.predict(X_test) 
print("Accuracy:",linear.score(X_test, y_test))
#Plotting the prediction
figure, ax = plt.subplots()
ax.scatter(y_test, predicted_linear)
ax.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title("Actual Truth vs Predicted")
plt.show()
```

```python
#Using Elastic Net
en=ElasticNet()
en.fit(X_train,y_train)
predicted_en = en.predict(X_test)
print("Accuracy:",en.score(X_test, y_test))
#Plotting the prediction
figure, ax = plt.subplots()
ax.scatter(y_test, predicted_en)
ax.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title("Actual Truth vs Predicted")
plt.show()
```

```python
#Using Ridge Regression
ridge=Ridge()
ridge.fit(X_train,y_train)
predicted_ridge = ridge.predict(X_test)
print("Accuracy:",ridge.score(X_test, y_test))
#Plotting the prediction
figure, ax = plt.subplots()
ax.scatter(y_test, predicted_ridge)
ax.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title("Actual Truth vs Predicted")
plt.show()
```

```python
#Using Lasso Regression
lasso=Lasso()
lasso.fit(X_train,y_train)
predicted_lasso = lasso.predict(X_test)
print("Accuracy:",lasso.score(X_test, y_test))
#Plotting the prediction
figure, ax = plt.subplots()
ax.scatter(y_test, predicted_lasso)
ax.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title("Actual Truth vs Predicted")
plt.show()
```

```python
#Using Random Forest Regression
forest=RandomForestRegressor()
forest.fit(X_train,y_train)
predicted_forest = forest.predict(X_test)
print("Accuracy:",forest.score(X_test, y_test))
#Plotting the prediction
figure, ax = plt.subplots()
ax.scatter(y_test, predicted_forest)
ax.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title("Actual Truth vs Predicted")
plt.show()
```

```python
#Using Decision Tree Regression
tree=DecisionTreeRegressor()
tree.fit(X_train,y_train)
predicted_tree = tree.predict(X_test)
print("Accuracy:",tree.score(X_test, y_test))
#Plotting the prediction
figure, ax = plt.subplots()
ax.scatter(y_test, predicted_tree)
ax.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title("Actual Truth vs Predicted")
plt.show()
```

```python
#Using Support Vector Regression
svr=SVR()
svr.fit(X_train,y_train)
predicted_svr = svr.predict(X_test)
print("Accuracy:",svr.score(X_test, y_test))
#Plotting the prediction
figure, ax = plt.subplots()
ax.scatter(y_test, predicted_svr)
ax.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title("Actual Truth vs Predicted")
plt.show()
```

```python
results = {'Regression_Model': ["Linear Regression","Elastic Net","Ridge Regression","Lasso Regression","Random Forest","Decision Tree","Support Vector"],
        'Accuracy': [0.886,0.887,0.887,0.887,0.987, 0.983,0.644]
          }
df = pd.DataFrame(results, columns = ['Regression_Model', 'Accuracy'])
print(df)
```

```python
print("Observations and Understanding")
print("1. Random Forest has the maximum accuracy.")
print("2. Support Vector has the worst accuracy")
print("3. The values that do not lie on the prediction line have more deviations and therfore we can say they have less accuracy")
print("4. With stronger correlation amongst the features TAX has higher accuracy whereas NoX have less accuracy as it less correlated")
print("Understanding on the models")
print("1. When there is a linear relationship between the variables then we use Linear Regression. To minimize the residual sum of squares between the observed targets, it fits a linear model with weight coefficient. Linear Approximation is used to predict the targets.")
print("2. L2 regularization is used in ridge Regression.To minimize a penalized residual sum of squares the ridge coefficients are used. The coeffients are more robust to collinearity.")
print("3. Coefficients are minimized to zero in Lasso Regression. It used L1 regularization. If there are more than one predictors it chooses one and reduces other to zero giving less number of features.")
print("4. It uses both L1 and L2 regularization.Elastic Net is used when variables are highly correlated and prone to double shrinkage.")
print("5. Decision Tree Algorithm is used for solving regression and classification problems. It is supervised learning algorithm.")
print("6. Random Forest Regression is a combination of decision trees. It reduces variance giving the best accuracy.")
print("7. Support vector regression is dependant on weight instances and therefore it is very noisy which reduces the accuracy. Changing the number of features can help to improve the accuracy.")
```
